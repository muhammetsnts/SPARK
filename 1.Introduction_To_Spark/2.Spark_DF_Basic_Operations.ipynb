```python
!wget -q https://raw.githubusercontent.com/muhammetsnts/SPARK/main/data/appl_stock.csv
```

```python
df = spark.read.csv("appl_stock.csv", inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.head(3)
```

```python
# filter stocks lower than 500$ using SQL 

df.filter("Close < 500").show()
```

```python
# We can grab a column 

df.filter("Close < 500").select("Open").show()
```

```python
df.filter("Close < 500").select(["Open", "Close"]).show()
```

```python
# Using python

df.filter(df["Close"] < 500).show()
```

### Filtering Multiple Conditions
 We should use `|` for OR, `&` for AND, `~` for NOT and use () for conditions. We cant use `and, or,not` directly.

```python
# and

df.filter((df.Close < 200) & (df.Open >200)).show()
```

```python
# not

df.filter((df.Close < 200) & ~(df.Open >200)).show()
```

```python
# equal

df.filter(df["Low"] == 197.16).show()
```

### collect() Method

```python
result = df.filter(df["Low"] == 197.16).collect()
result
```

```python
result[0]
```

```python
row = result[0]
```
